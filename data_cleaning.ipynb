```python
import pandas as pd
```

```python
courses = pd.read_csv(open('courses.csv', encoding='latin'))
q = pd.read_csv(open('courses_q.csv', encoding='latin'))
```

```python
full_courses = pd.merge(courses, q, on='cat_num')
full_courses.set_index('course_id')
full_courses.drop_duplicates(subset=['number'])
```

```python
course_table = full_courses[['title', 'description', 'department_id', 'number', 'course_id']]
course_table.set_index('course_id')
course_table = course_table.drop_duplicates(subset=['title'])
```

```python
course_table.to_csv(open('course_table.csv', 'w'))
```
